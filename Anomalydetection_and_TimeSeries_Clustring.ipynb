```python
'''
Question 1 : What is Dimensionality Reduction? Why is it important in machine learning?
Answer:      Dimensionality reduction is the process of reducing the number of input variables (features) in a dataset
             while preserving as much relevant information as possible. In machine learning, it helps simplify complex
             data by projecting it into a lower-dimensional space.
             
             It is important because high-dimensional data can lead to problems like overfitting, increased computation time,
             and difficulty in visualization. By reducing dimensions, we improve model performance, make data easier to interpret,
             and enhance the efficiency of algorithms. Techniques like PCA or t-SNE help reveal hidden patterns by eliminating 
             redundant or less informative features.

Question 2: Name and briefly describe three common dimensionality reduction techniques..
Answer:  a. Principal Component Analysis (PCA):
            PCA is a linear technique that transforms the original variables into a new set of uncorrelated variables called
            principal components. These components capture the maximum variance in the data, helping reduce dimensionality
            while preserving essential information.

         b. Linear Discriminant Analysis (LDA):
            LDA is a supervised technique used to find a linear combination of features that best separates two or more classes.
            It reduces dimensions while maximizing class separability, making it useful in classification tasks.

         c. t-Distributed Stochastic Neighbor Embedding (t-SNE):
            t-SNE is a nonlinear method mainly used for visualization of high-dimensional data. It reduces dimensions by preserving
            local relationships between data points, often used for clustering and visual interpretation.

Question 3: What is clustering in unsupervised learning? Mention three popular clustering algorithms.
Answer:     Clustering is an unsupervised learning technique used to group similar data points together based on their features or patterns,
            without using predefined labels. The goal is to ensure that data points within the same cluster are more similar to each other than
            to those in other clusters.

            Three popular clustering algorithms are:

         a. K-Means Clustering: Divides data into k clusters by minimizing the variance within each cluster.
         b. Hierarchical Clustering: Builds a tree-like structure of clusters (dendrogram) by merging or splitting them based on similarity.
         c. DBSCAN (Density-Based Spatial Clustering of Applications with Noise): Groups data points based on density, effectively identifying
            clusters of arbitrary shape and detecting outliers.       

Question 4: Explain the concept of anomaly detection and its significance.
Answer:     Anomaly detection involves identifying data points or patterns that significantly deviate from normal behavior. These unusual
            observations, known as anomalies or outliers, can indicate critical events such as fraud, system faults, or network intrusions.
            It is significant because detecting anomalies early helps organizations prevent losses, maintain system reliability, and improve
            decision-making by ensuring data integrity and safety.            

Question 5: List and briefly describe three types of anomaly detection techniques.
Answer:  a. Statistical Methods: Use probability distributions or statistical tests to identify data points that deviate from expected behavior.
         b. Machine Learning Methods: Apply algorithms like Isolation Forest or One-Class SVM to learn normal patterns and detect deviations.
         c. Distance/Clustering-Based Methods: Detect anomalies based on their distance from cluster centers or density of surrounding points 
            (e.g., DBSCAN).

Question 6: What is time series analysis? Mention two key components of time series data.
Answer:     Time series analysis involves studying data points collected or recorded at specific time intervals to identify patterns,
            trends, and dependencies over time.
            
            Two key components of time series data are:

         a. Trend: The long-term increase or decrease in the data over time.
         b. Seasonality: The repeating short-term pattern or cycle that occurs at regular intervals (e.g., monthly, quarterly).  

Question 7: Describe the difference between seasonality and cyclic behavior in time series.
Answer:     Seasonality refers to regular and predictable patterns that repeat over a fixed period, such as daily, monthly,
            or yearly (e.g., higher ice cream sales in summer).
          
            Cyclic behavior, on the other hand, occurs over irregular periods, often influenced by economic or external factors
            (e.g., business cycles). The main difference is that seasonality is time-dependent and fixed, while cycles are variable in duration.

'''                 
```

```python
#Question 8: Write Python code to perform K-means clustering on a sample dataset.
#Answer:

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=200, centers=3, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            color='red', marker='x', s=100, label='Centroids')
plt.title("K-Means Clustering Example")
plt.legend()
plt.show()
```
